- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh 


```python
n = float(input(" So kwh dien tieu thu la: "))
if 0<=n<=50:
    tong_tien = n*1500
elif 51<=n<=100:
    tong_tien = 50*1500 + (n-50)*2000
else:
    tong_tien = 50*1500 + 50*2000 + (n-100)*3000
print(tong_tien)
```
