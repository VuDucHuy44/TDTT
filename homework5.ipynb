W5A3: Kiểm tra số nguyên tố

```python
import math

def is_prime(n: int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

W5A8: Tính khoảng cách Hamming giữa hai số nguyên

```python
def sum_of_digits(n: int) -> int:
    total = 0
    n = abs(n) # Đảm bảo số dương
    while n > 0:
        total += n % 10  # Lấy chữ số cuối cùng (chia lấy dư cho 10)
        n //= 10         # Loại bỏ chữ số cuối cùng (chia nguyên cho 10)
    return total
```

W5A10: Kiểm tra hai từ đẳng cấu (Isomorphic Strings)

```python
def are_isomorphic(a: str, b: str) -> bool:
    # Giả sử hai chuỗi có cùng độ dài như đề bài nêu
    mapping = {}
    mapped_chars = set() # Để đảm bảo không có 2 ký tự khác nhau cùng ánh xạ về 1 ký tự
    
    for char_a, char_b in zip(a, b):
        if char_a in mapping:
            # Nếu char_a đã có ánh xạ rồi, nó phải khớp với char_b hiện tại
            if mapping[char_a] != char_b:
                return False
        else:
            # Nếu char_a chưa có ánh xạ, nhưng char_b đã bị ký tự khác chiếm dụng
            if char_b in mapped_chars:
                return False
            # Thiết lập ánh xạ mới
            mapping[char_a] = char_b
            mapped_chars.add(char_b)
            
    return True
```

W5A12: Chia lấy nguyên và dư (Divide with Remainder)

```python
from typing import Tuple

def divide_with_remainder(dividend: int, divisor: int) -> Tuple[int, int]:
    # Kiểm tra lỗi chia cho 0
    if divisor == 0:
        return (0, 0) # Hoặc xử lý lỗi tùy yêu cầu
        
    # Tính thương (q) và số dư (r)
    q = dividend // divisor
    r = dividend % divisor
    
    # Trả về một bộ (tuple) gồm q và r
    return (q, r)
```

W5A21 Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def is_one_product_of_others(a,b,c):
    dieu_kien_1 = (a==b*c)
    dieu_kien_2 = (b==a*c)
    dieu_kien_3 = (c==a*b)
    return (dieu_kien_1) or (dieu_kien_2) or (dieu_kien_3)
print(f"Kiem tra(6,2,3): {is_one_product_of_others(6,2,3)}")
print(f"Kiểm tra (5, 30, 6): {is_one_product_of_others(5, 30, 6)}")
print(f"Kiểm tra (2, 2, 5): {is_one_product_of_others(2, 2, 5)}")
```

```python
import math

def tinh_goc_abc(x1, y1, x2, y2, x3, y3):
    ba_x, ba_y = x1 - x2, y1 - y2
    bc_x, bc_y = x3 - x2, y3 - y2
    tich_vo_huong = ba_x * bc_x + ba_y * bc_y
    do_dai_ba = math.sqrt(ba_x**2 + ba_y**2)
    do_dai_bc = math.sqrt(bc_x**2 + bc_y**2)
    cos_phi = tich_vo_huong / (do_dai_ba * do_dai_bc)
    cos_phi = max(-1, min(1, cos_phi))
    
    goc_radian = math.acos(cos_phi)
    return math.degrees(goc_radian)
```
