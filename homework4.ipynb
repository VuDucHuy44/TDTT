"[LargestPrimeFactor]
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 

```python
n = int(input("Nhập vào số nguyên dương n (n >= 2): "))
tam = n
uoc_lon_nhat = -1
while n % 2 == 0:
    uoc_lon_nhat = 2
    n //= 2
i = 3
while i * i <= n:
    while n % i == 0:
        uoc_lon_nhat = i
        n //= i
    i += 2
if n > 1:
    uoc_lon_nhat = n
print(f"Ước số nguyên tố lớn nhất của {tam} là: {uoc_lon_nhat}")
```
